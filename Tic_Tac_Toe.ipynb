```python
import random

def print_board(board):
    for i in range(3):
        print(" | ".join(board[i]))
        if i < 2:
            print("---------")

def check_win(board, symbol):
    # Rows
    for row in board:
        if row == [symbol, symbol, symbol]:
            return True

    # Columns
    for col in range(3):
        if board[0][col] == symbol and board[1][col] == symbol and board[2][col] == symbol:
            return True

    # Diagonals
    if board[0][0] == symbol and board[1][1] == symbol and board[2][2] == symbol:
        return True
    if board[0][2] == symbol and board[1][1] == symbol and board[2][0] == symbol:
        return True

    return False

def board_full(board):
    for row in board:
        if " " in row:
            return False
    return True

def get_valid_index(prompt):
    while True:
        value = input(prompt)
        if value.isdigit() and int(value) in [0, 1, 2]:
            return int(value)
        print("Invalid input. Choose 0, 1, or 2.")

def computer_move(board, symbol):
    empty_cells = [(r, c) for r in range(3) for c in range(3) if board[r][c] == " "]
    if empty_cells:
        r, c = random.choice(empty_cells)
        board[r][c] = symbol
        print(f"Computer placed {symbol} at ({r}, {c})")

# --- Replay loop ---
while True:

    # Reset board for each new game
    board = [
        [" ", " ", " "],
        [" ", " ", " "],
        [" ", " ", " "]
    ]

    player_symbol = input("Choose X or O: ").upper()
    computer_symbol = "O" if player_symbol == "X" else "X"

    print("\nGame start!")
    print_board(board)

    # --- One full game ---
    while True:
        print("\nYour turn:")
        while True:
            row = get_valid_index("Row (0–2): ")
            col = get_valid_index("Col (0–2): ")
            if board[row][col] == " ":
                board[row][col] = player_symbol
                break
            print("That cell is taken.")

        print_board(board)

        if check_win(board, player_symbol):
            print("You win!")
            break

        if board_full(board):
            print("It's a draw!")
            break

        print("\nComputer's turn:")
        computer_move(board, computer_symbol)
        print_board(board)

        if check_win(board, computer_symbol):
            print("Computer wins!")
            break

        if board_full(board):
            print("It's a draw!")
            break

    # --- Ask for replay ---
    again = input("\nPlay again? (y/n): ").lower()
    if again != "y":
        print("Thanks for playing!")
        break
```
